```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "AcNo": ["A00023", "A00121", "A00234", "A01208", "A01208", "A00789", "A00893", "A00987"],
    "CustID": ["C02", "C01", "C02", "C06", "C01", "C04", "C05", "C03"],
    "Type": ["Savings", "Current", "Current", "Current", "Current", "Savings", "Savings", "Current"],
    "BranchID": ["B04", "B01", "B04", "B02", "B02", "B04", "B06", "B02"],
    "Balance": [90000, 80567, 12000, 45000, 45000, 4500, 67900, 78600]
    }


df_account = pd.DataFrame(data)
```

```python
# ii. Print the second and third columns
print("\nSecond and Third Columns:")
print(df_account.iloc[:, 1:3])
```

```python
# iii. Print the 4th to 7th rows
print("\n4th to 7th Rows:")
print(df_account.iloc[3:7])
```

```python
# iv. Print the column names
print("\nColumn Names:")
print(df_account.columns)
```

```python
# v. Print records where balance is between 30k to 40k
print("\nRecords with Balance between 30K to 40K:")
print(df_account[(df_account["Balance"] >= 30000) & (df_account["Balance"] <= 40000)])
```

```python
# vi. Print records where type is "Savings" and Branch is "B04"
print("\nRecords where Type is 'Savings' and Branch is 'B04':")
print(df_account[(df_account["Type"] == "Savings") & (df_account["BranchID"] == "B04")])
```

```python

# vii. Print the average Balance Branchwise
print("\nAverage Balance per Branch:")
print(df_account.groupby("BranchID")["Balance"].mean())
```

```python
# viii. Plot a bar plot showing the number of Accounts per Branch
df_account["BranchID"].value_counts().plot(kind="bar", color="skyblue", edgecolor="black")
plt.xlabel("Branch ID")
plt.ylabel("Number of Accounts")
plt.title("Number of Accounts per Branch")
plt.show()
```
