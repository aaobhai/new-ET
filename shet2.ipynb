```python
# Accept inputs from the user
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Perform arithmetic operations
sum_result = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2 if num2 != 0 else "Undefined (division by zero)"
modulus = num1 % num2 if num2 != 0 else "Undefined (modulus by zero)"

# Display results
print("\nResults of Arithmetic Operations:")
print(f"Sum: {sum_result}")
print(f"Difference: {difference}")
print(f"Product: {product}")
print(f"Quotient: {quotient}")
print(f"Modulus: {modulus}")
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

# Accept input from user
num = int(input("Enter a number: "))

# Calculate and print factorial
if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    print(f"The factorial of {num} is: {factorial(num)}")
```

```python

def sum_of_digits(n):
    total = 0
    while n > 0:
        total += n % 10  # Get the last digit and add to total
        n //= 10         # Remove the last digit
    return total

# Accept input from the user
num = int(input("Enter a number: "))

# Ensure the number is positive
num = abs(num)

# Calculate and print the sum of digits
print(f"The sum of digits of {num} is: {sum_of_digits(num)}")
```

```python
import math

# Lambda function to calculate distance
distance = lambda x1, y1, x2, y2: math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)

# Accept inputs from the user
x1, y1 = map(float, input("Enter coordinates of first point (x1 y1): ").split())
x2, y2 = map(float, input("Enter coordinates of second point (x2 y2): ").split())

# Calculate and print the distance
print(f"Distance between points ({x1}, {y1}) and ({x2}, {y2}) is: {distance(x1, y1, x2, y2):.2f}")

```

```python
# Accept input from the user
num = int(input("Enter a number: "))

# Using while loop to check if the number is even or odd
while True:
    if num % 2 == 0:
        print(f"{num} is an Even number.")
    else:
        print(f"{num} is an Odd number.")
    break  # Exit the loop after checking
```

```python
def dynsum(*numbers):
    return sum(numbers)  # Calculate sum of all numbers

# Accept input from the user as space-separated values
nums = list(map(float, input("Enter numbers separated by space: ").split()))

# Call the function and print the result
print(f"The sum of the numbers is: {dynsum(*nums)}")

```

```python
# Creating sets A and B
A = {10, 15, 34, 89, 90}
B = {2, 4, 6, 10, 15, 30, 40, 90}

# Union of A and B
union_result = A | B
print("Union of A and B:", union_result)

# Intersection of A and B
intersection_result = A & B
print("Intersection of A and B:", intersection_result)

# Difference of A and B (A - B)
difference_result_A_B = A - B
print("Difference of A and B (A - B):", difference_result_A_B)

# Difference of B and A (B - A)
difference_result_B_A = B - A
print("Difference of B and A (B - A):", difference_result_B_A)

# Symmetric Difference of A and B
symmetric_difference_result = A ^ B
print("Symmetric Difference of A and B:", symmetric_difference_result)
```

```python
 # Accepting 10 numbers from the user
numbers = []
for i in range(10):
    num = float(input(f"Enter number {i+1}: "))
    numbers.append(num)

# Calculating the average
average = sum(numbers) / len(numbers)
print(f"The average of the entered numbers is: {average}")
```
