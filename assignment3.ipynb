```python
import pandas as pd

# Creating the DataFrame
data = {
    "CustID": ["C01", "C02", "C03", "C04", "C05", "C06", "C07", "C08"],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hank"],
    "Age": [25, 30, 35, 40, 45, 50, 28, 33],
    "Education": [1, 2, 3, 2, 1, 3, 2, 1],
    "Balance": [80000, 95000, 120000, 70000, 110000, 92000, 85000, 100000]
}

df_cust = pd.DataFrame(data)
```

```python
# ii. Print the second to fourth columns
print("\nSecond to Fourth Columns:")
print(df_cust.iloc[:, 1:4])
```

```python
# iii. Print the 3rd and 5th rows
print("\n3rd and 5th Rows:")
print(df_cust.iloc[[2, 4]])
```

```python
# iv. Print the column names
print("\nColumn Names:")
print(df_cust.columns)
```

```python

# v. Print records where balance is greater than 90000
print("\nRecords where Balance is greater than 90000:")
print(df_cust[df_cust["Balance"] > 90000])
```

```python
# vi. Print records where Education is 2
print("\nRecords where Education is 2:")
print(df_cust[df_cust["Education"] == 2])
```

```python
# vii. Print the average balance as per age
print("\nAverage Balance per Age:")
print(df_cust.groupby("Age")["Balance"].mean())
```

```python
# viii. Print the maximum balance as per education type
print("\nMaximum Balance per Education Type:")
print(df_cust.groupby("Education")["Balance"].max())
```
